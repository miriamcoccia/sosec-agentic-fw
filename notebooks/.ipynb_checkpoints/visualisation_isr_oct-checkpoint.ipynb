## October 2023: Timeline of Telegram Posts about Israel

This notebook explores the possible changes in the amount of Telegram posts about Israel during the month of October 2023, which was characterised by Israel's declaration of war against Hamas after the organization's attack at the Nova music festival on the 7th October. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.widgets import Cursor
import glob
import os
```

```python
%matplotlib inline
```

```python
raw_data = pd.read_csv(r"../data/post_isr_result.csv", usecols=["Post Text", "date", "About Israel"])
# only posts about israel, e.g. where the value in the "About Israel" is 1
data = raw_data[raw_data["About Israel"] == "1"].copy()
```

```python
print(f"Percentage of Telegram posts about Israel from the 1st October 2023 to the 31st October 2023: \n{int(round(len(data) / len(raw_data), 2) * 100)}%")
```

#### Datatypes
Adjusting the datatypes of the columns for further analysis

```python
raw_data["About Israel"] = pd.to_numeric(raw_data["About Israel"], errors="coerce")
```

```python
raw_data["date"] = pd.to_datetime(raw_data["date"])
```

```python
raw_data = raw_data.set_index("date")
# sort chronologically
raw_data = raw_data.sort_index()
```

### Distribution of Posts about Israel

```python
fig = plt.subplots(figsize=(16, 5))
plt.plot(raw_data.index, raw_data["About Israel"])
plt.title("Posts about Israel October 2023", fontsize=20)
plt.xlabel("Date", fontsize=15)
plt.ylabel("Nr. Posts", fontsize=15)
plt.xticks(rotation=30, fontsize=10)
plt.show();
```

### Distribution of World Events about Israel

```python
# Combining all event files into one dataframe. Why are there so many files? Because we used several keywords to query ER and extract the events.
EVENTS_FOLDER = r"../data/isr_events"
all_event_files = glob.glob(os.path.join(EVENTS_FOLDER, "*.parquet"))
all_event_dfs = [pd.read_parquet(file) for file in all_event_files]
events_data = pd.concat(all_event_dfs, ignore_index=True) 
events_data = events_data[["eventDate", "totalArticleCount", "title.eng"]]
```

```python
events_data["totalArticleCount"] = pd.to_numeric(events_data["totalArticleCount"], errors="coerce")
events_data["eventDate"] = pd.to_datetime(events_data["eventDate"])
events_data = events_data.set_index("eventDate")
events_data = events_data.sort_index()
```

```python
fig, ax = plt.subplots(figsize=(16, 5))
sc = ax.scatter(events_data.index, events_data["totalArticleCount"], picker=True)
ax.set_title("Events About Israel October 2023", fontsize=20)
ax.set_xlabel("Date", fontsize=15)
ax.set_ylabel("Nr. Articles", fontsize=15)
ax.tick_params(axis="x", rotation=30, labelsize=10)

annot = ax.annotate("", xy=(0, 0), xytext=(10, 10), textcoords="offset points",
                    bbox=dict(boxstyle="round", fc="w"), arrowprops=dict(arrowstyle="->"))
annot.set_visible(False)

cursor_line = ax.axvline(color='red', linewidth=1, linestyle="--")
def update_annot(ind):
    """
    Update annotation position and text.
    """
    pos = sc.get_offset()[ind[0]]
    annot.xy = pos
    title = events_data.iloc[ind[0]]["title.eng"]
    annot.set_text(title)
    annot.get_bbox_patch().set_alpha(0.7)

def on_hover(event):
    """
    Display annotation on hover.
    """
    vis = annot.get_visible()
    if event.inaxes == ax:
        cont, ind = sc.contains(event)
        if cont:
            update_annot(ind["ind"])
            annot.set_visible(True)
            fig.canvas.draw_idle()
        else:
            if vis:
                annot.set_visible(False)
                fig.canvas.draw_idle()
fig.canvas.mpl_connect("motion_notify_event", on_hover)
plt.show();
```
